```python
# Import package
from docplex.mp.model import Model
```

```python
# create one model instance, with a name
m = Model(name='Inform')
```

```python
# qty produce (number of unit)
qty_produce = { (s,c,p): m.integer_var(name= '{0} {1} {2}'.format(s,c,p))  for s in site for c in color for p in packing }
```

```python
# Sum of output (in pounds) <= input

m.add_constraints(qty_produce[s,c,'Box']*2.5 + qty_produce[s,c,'Bag']*0.25  <= rmi_dict.get((s,c),0) for s in site for c in color ) 
```

```python
# Total Production must Meet demand
m.add_constraints( m.sum(qty_produce[(s,c,p)] for s in site)  >= order_dict[(c,p)] for c in color for p in packing)
```
